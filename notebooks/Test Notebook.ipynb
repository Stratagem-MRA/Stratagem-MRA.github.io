The first step for this project is to identify an optimal $\omega$ for the SOR algorithm. According to https://crunchingnumbers.live/2017/07/09/iterative-methods-part-2/ the optimal choice for our problem is given by $\omega=\frac{2}{1+sin(h\pi)}$ where h is the distance between mesh points. Given that $h=\frac{1}{N+1}=\frac{1}{51}$ we can compute that the optimal $\omega\approx 1.8840$
